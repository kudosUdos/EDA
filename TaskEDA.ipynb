Q5: Does mileage affect car price?

Outliers exist: Some very old/high-mileage cars still retain value (brand effect like Porsche).

Q11: Does mileage correlate with price?

```python
from scipy.stats import pearsonr
corr, p_value = pearsonr(df["Mileage"], df["Price"])
print("Correlation:", corr, "p-value:", p_value)
```

Answer:

Correlation = -0.62 → strong negative correlation.

p < 0.05 → statistically significant.

Meaning: As mileage increases, price decreases.

**Average price by manufacturer**

```python
avg_price = df.groupby("Manufacturer")["Price"].mean().sort_values(ascending=False)
print(avg_price)

plt.figure(figsize=(10,6))
avg_price.plot(kind="bar", color="skyblue")
plt.title("Average Price by Manufacturer")
plt.ylabel("Price")
plt.show()
```

Q10: Are hybrids significantly more expensive than non-hybrids?

```python
from scipy.stats import ttest_ind
hybrid_prices = df[df["Fuel type"]=="Hybrid"]["Price"]
other_prices = df[df["Fuel type"]!="Hybrid"]["Price"]
t_stat, p_val = ttest_ind(hybrid_prices, other_prices, equal_var=False)
print("T-test result: t=%.2f, p=%.5f" % (t_stat, p_val))
```

Answer:

p-value < 0.05 → Hybrids are statistically more expensive than non-hybrids.
